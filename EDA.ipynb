Date colomundaki tarih tipini degistirmek istiyorum. Cünkü evin satildigi andaki yasini hesaplama yapmak icin year kolonundan yararlanabilirim. Bu yüzden dataframe'mize year_sold kolonu ekledim. Bunun yaninda sectigim client'ta yilin belirli zamanlari icin herhangi bir karakteristik özelligi olmadigindan sonrasinda date kolonunu df'ten silebilirim. Cünkü artik isime yaramayacak. Bunun yaninda id.1 kolununa da ihtiyacim yok. Cünkü su an herhangi bir anlam ifade etmiyor.

```python
# we want to preprcess the `data` and extract the year in order to calcuate the age of the house.
df_houses['year_sold'] = pd.to_datetime(df_houses['date']).dt.date.astype('datetime64[ns]')
df_houses['year_sold'] = pd.DatetimeIndex(df_houses['year_sold']).year # extract only year
```

```python
df_houses.drop(['id.1', 'date', 'house_id'], axis = 1, inplace= True)
```

```python
df_houses.head()
```

Burada da evin satildigi andaki yasini hesaplama yapalim ve df'ye ekleyelim.

```python
# We want to know about house age when it sold.
df_houses['house_age'] = df_houses['year_sold'] - df_houses['yr_built']
df_houses['house_age']
```

Grafikte de görüldügü gibi 'waterfront' kolonunda yaklasik %10'a yakin bir kayip veri var. 'yr_renovated' kolonunda ise yaklasik %20 kayip veri söz konusu.
```python
print(f"numbers of rows : {df_houses.shape[0]}")
print(f"missing values in waterfront : {round(df_houses.waterfront.isna().sum()/df_houses.shape[0]*100,2)} %")
print(f"missing values in yr_renovated : {round(df_houses.yr_renovated.isna().sum()/df_houses.shape[0]*100,2)} %")
print(f"missing values in view : {round(df_houses.view.isna().sum()/df_houses.shape[0]*100,2)} %")
print(f"missing values in sqft_basement : {round(df_houses.sqft_basement.isna().sum()/df_houses.shape[0]*100,2)} %")
print(f"missing values in data frame : {round(df_houses.isna().sum().sum()/(df_houses.shape[0]*df_houses.shape[1])*100,2)} %")
```

Data frame'in icindeki 1.42% missing value degeri kabul edilebilir. Ancak kolonlarin icindeki missing value'lara baktigimizda 'waterfront' da 11.07% , 'yr_renovated' da ise 17.82% gibi büyük bir kayip söz konusu. Bu yüzden buradaki missing value'lari atmak yerine doldurmaya calisabiliriz.

'View' ile 'sqft_basement' kolonundaki missing value'lari atabiliriz. Cünkü bizim analizimizi etkileyecek kadar fazla deger yok.

Ancak bunun icin bu verilerin girilmeme sebepleri üzerinde düsünmemiz gerekir. Bu missing value'lari incelerken bu kolonla iliski diger özelliklere de bakmak gerekebilirr ve anlamli cikarimlar cikarilabilir.

Missing valuelari doldururken interpolation kullanamayiz. Cünkü 'waterfront' kolonu kategorik data tipindedir. Burada 0 veya 1 vermemiz hedef kolonu icin önemli degisikliklere sebep olabilir. Cünkü evin fiyati belirlenirken sahilde olup olamamasi ve renovasyon yilinin erken veya gec olmasi önemli bir faktör olabilir.

Ancak modeli fazla bozmamak icin 'waterfront' ve 'yr_renovated' kolonlarinin median valuelari ile missing valueleri doldurabiliriz.

```python
# First, copy the actual data frame for filling missing values
df_houses_copy = df_houses.copy()
```

```python
df_houses_copy.fillna({
    "waterfront" : df_houses_copy['waterfront'].median(),
    "yr_renovated" : df_houses_copy['yr_renovated'].median()
}, inplace = True)

df_houses_copy.dropna(inplace=True, ignore_index=True)
```

```python
df_houses_copy.count()
```

Böylelikle missing value olmayan bir data frame elde etmis olduk. Simdi bu df üzerinden calisabiliriz.

```python
df_houses_copy.info()
```

```python
# Numerical data distribution 

df_houses_num = df_houses_copy.select_dtypes(include = ['float64', 'int64'])
df_houses_num.head()
```

```python
# ID kolonuna ihtiyacimiz olmadigi icin onu da df'ten cikarabiliriz.
df_houses_num.drop(['id'], axis = 1, inplace=True)
```

```python
# df_houses_num['yr_renovated'] = (df_houses_num['yr_renovated'] / 10).astype('int64')
df_houses_num[['sqft_living', 'sqft_lot', 'floors', 'waterfront', 'view', 'sqft_above', 'sqft_basement', 'sqft_living15', 'sqft_lot15', 'price']] = df_houses_num[['sqft_living', 'sqft_lot', 'floors', 'waterfront', 'view', 'sqft_above', 'sqft_basement', 'sqft_living15', 'sqft_lot15', 'price']].astype('int64')

df_houses_num.head()
```

Simdi ise buradaki özelliklerin price ile olan correlation'larina bakalim.

```python
df_houses_num_corr = df_houses_num.corr(method='spearman')['price'][:]
str_corr_list = df_houses_num_corr[abs(df_houses_num_corr) > 0.5].sort_values(ascending=False)
print("There is {} strongly correlated values with Price:\n{}".format(len(str_corr_list), str_corr_list))
```

Perfect, we now have a list of strongly correlated values but this list is incomplete as we know that correlation is affected by outliers. So we could proceed as follow:

Plot the numerical features and see which ones have very few or explainable outliers
Remove the outliers from these features and see which one can have a good correlation without their outliers
Btw, correlation by itself does not always explain the relationship between data so ploting them could even lead us to new insights and in the same manner, check that our correlated values have a linear relationship to the SalePrice.

For example, relationships such as curvilinear relationship cannot be guessed just by looking at the correlation value so lets take the features we excluded from our correlation table and plot them to see if they show some kind of pattern.

```python
for i in range(0, len(df_houses_num.columns), 5):
    sns.pairplot(data=df_houses_num,
                x_vars=df_houses_num.columns[i:i+5],
                y_vars=['price'])
```
